### split( ) : 將字串進行切割。

str.split(分割符號)

str.split(分割符號,分割次數)


```python
str4 = "python-java-c-ruby-VB"
print(str4.split("-"))#左往右割
print(str4.split("-",1))
print(str4.split("-",2))
print(str4.split("-",0))
print(str4.split("-",-1))

print(str4.rsplit("-")) #右往左割
print(str4.rsplit("-",1))
print(str4.rsplit("-",2))
```

## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。

```python
list1 = ["H","i"]
tuple2 = ('H','i')
print(type(list1),list1)
print(type(tuple2),tuple2)
print("len:" ,len(list1))
list1[1] = 'a'
print(list1)
```

```python
list1 = ["a","b","c","a","d","c"]
print(list1)
mySet = set(list1) #list 轉 set
print(mySet)
list2 = list(mySet) # set轉list
print(list2)
tuple2 = tuple(list2)# list轉tuple
print(tuple2)
```

## list指派拷貝與附加

### list可以新增資料：

+ 可以用append( )方式附加資料，代表這一組資料成為一個資料。
+ 可以用extend( )方式擴展資料，代表這一組資料是一群資料。

```python
list1 = ['a','b','c']
print("附加前:",list1)
list1.append("def")
print("附加後:",list1)
list2 = ["def","ghij"]
list1.append(list2)
print("附加後:",list1)#沒有展開
list2 = ["x","y"]
#list1 = list1 + list2
print(list1 + list2)#合併產生新的list
print(list1)
```

## list 項目擴展資料
+ extend 擴展資料：
  + 如果引號括起來的字串有 [ ] 將會拆解成好幾個資料。
  + 可以把其他 list 加入到這個 list 內，擴展多個欄位。

```python
list1 = ['a','b','c']
print("擴展前:",list1)
list1.extend(["def","ghij"])
print("擴展之後:",list1)
#list1 += ["def","ghij"]
```

## list指派拷貝!

```python
list1 = [1,2,3,4,5,6.6,"list1"]
list2 = [11,12,13,13,14,16.6,"list2"]
print("list1:",id(list1),"list2:",id(list2))
list3 = list1 #淺複製
print(list1,list3)
print("list1:",id(list1),"list3:",id(list3))

list4 = list1.copy()
print(list1,list4)
print("list1:",id(list1),"list4:",id(list4))

list5 = ["a","b","c"]
list6 = ["123","456"]
print("list5:",id(list5))
print("list5+之前:",list5)
list5 = list5 + list6 #產生新的list
print("list5:",id(list5))
print("list5+之後:",list5)

print("list5:",id(list5))
print("list5+=之前:",list5)
list5 += list6 #等同於extend
print("list5:",id(list5))
print("list5+=之後:",list5)

list7 = ["a","b","c"]
list8 = ["123","456"]
print(id(list7))
list7.extend(list8)
print(list7)
print(id(list7))
```
