## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = 'one, two...ten'
re1  = re.split("[,. ]+",text)
print(re1)
re2  = re.split("[,. ]+",text,maxsplit=1)
print(re2)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。


```python
import re
inputStr = "hello 123 world 456"
rStr = re.sub('\d+',"222",inputStr)
print(rStr)
rStr = re.sub('\d+',"222",inputStr,count = 1)
print(rStr)
```

## __init__(self, …) 初始化方法

```python
class Car:
    def __init__(self,color,seat):
        print(self)
        self.color = color
        self.seat = seat
    def printInfo(self):
        print(self.color)
        print(self.seat)
car1 = Car("blue",4)
print(car1.color)
print(car1.seat)
car2 = Car("Red",2)
car2.printInfo()
car3 = Car("Black",3)
car3.printInfo()
```

## 類別方法(Class Method)


```python
class Student:
    company = "GJ"
    phone = "8825252"
    def __init__(self,name):
        self.name = name
    def printInfo(self):    
        #print("self:",self)
        print("name:",self.name)
     #類別函數
    @staticmethod
    def getCompany():
        return Student.company
    
    @classmethod
    def getPhone(cls): 
        print("cls:",cls)
        return "Phone:"+cls.phone
    #建立物件
    @classmethod
    def newStudent(cls,name):
        return cls(name)
    
st1 = Student("Ken")    
st1.printInfo()
Student.getCompany()
print(st1.getPhone())

st2 =  Student.newStudent("Vivin")
st2.printInfo()
#print(st1.getCompany())
```

```python
class Person:
    #類別屬性共用
    count = 0
    def __init__(self,name):
        Person.count += 1
        self.name = name
    def printInfo(self):
        print(self.name,f"有{Person.count}位Person")
p1 = Person("Ken")
p2 = Person("Vivin")
#p3 = Person("Lucy")

p1.printInfo()#Ken 有2位Person
p2.printInfo()#Vivin 有2位Person
```

## 封裝
屬性與方法加上__為私有

```python
class Person:
    def __init__(self,name,age):
        self.__name = name
        self.__age = age
    def printInfo(self):
        print(self.__name,self.__age)
    def setAge(self,age): 
        if age < 0 or age > 200:
            raise OverflowError("錯誤的年齡!")
        self.__age = age
    def getAge(self):
        return self.__age
p1 = Person("Ken",15)        
p1.printInfo()
#p1.__age = -10000
p1.setAge(100)
print(p1.getAge())
p1.printInfo()
```
