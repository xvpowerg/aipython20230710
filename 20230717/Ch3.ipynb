## while 迴圈

當 while 迴圈的條件分析成立時，才會進入到迴圈。

若 while 迴圈的條件分析不成立，則跳出迴圈。

若 while 迴圈內條件分析一開始不成立，則迴圈一次都不執行。


```python
a = 6
while a > 0:
    print(a)
    a -= 1
print("離開後的a:",a)    
```

```python
a = 15
while a > 0:
    print(a)
    a -= 2
    
```

## 迴圈就是要從頭跑到尾嗎?迴圈可以做中斷。

所謂的中斷就是執行到你要中斷的那一行…

那…然後呢？有兩種選擇：

break 代表不回頭，跳出迴圈。

continue 代表繼續執行下一個迴圈。

要挑選哪一個就依照你中斷之後的下一步來做決定。

```python
for i in range(1,11):    
    if i == 5:
        break
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```
