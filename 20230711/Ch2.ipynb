## FOR迴圈
### 知道數量時使用for loop
### python的for 有一種由頭到取數值的想法


```python
#預設請況產生的數字>=0 < 5
for i in range(5):
    print(i)
    
```

```python
# 產生的數字>=2 < 8
for k in range(2,8):
    print(k)
```

```python
for x in range(2,18,3):
    print(x)
```

```python
for y in range(10,1,-2):
    print(y)
```

```python
for i in range(4):
    print(i)
print("離開的i:",i)  
for i in range(-4):
    print(i)
print("離開的i:",i)      
```

```python
for i in range(4,8):
    print(i)
print("1 離開後的i:",i)
for i in range(-4,-8):
    print(i)
print("2 離開後的i:",i)
for i in range(8,4):
    print(i)
print("3 離開後的i:",i)
for i in range(-8,-4):
    print(i)
print("4 離開後的i:",i)    
```

## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈。


```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
print("End")    
```

```python
for i in range(1,11):
    if i == 5 or i == 7:
        continue
    print(i)
print("End")    
```

```python
print("請輸入大於80的整數")
while True:
    a = int(input("請輸入數字:"))
    if a > 80:
        print("a=",a)
        break
print("恭喜跳離迴圈")        
```
