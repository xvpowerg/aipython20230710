## Sort
+ list 進行排序有 sorted( ) 與 sort( ) 兩個動作。
+ 兩者差異：
  + sorted()
    + 不會影響 list 本身結構，且可以輸出結果。
    + sorted(list項目)。
    + sorted( ) 可用來排序任何項目。
+ sort()
    + 會影響 list 本身結構，且不可以於操作這個動作時進行輸出。
    + list.sort( )。
    + sort( ) 只能用在list上排序。

### 排序原則與參數
+ 排序原則：
  + 數字則依照大小排序。
  + 字串則依照 ASCII 編碼順序進行排序。
+ 可加入的參數：
   + reverse=True  
     + 有這個參數由大到小排序
     + 若沒有這個參數則由小到大排序
+ key=len
     + 依照 list 內元素字串長度進行排序
+ key=str.upper
     + 將 list 內元素轉換為大寫，再依照 ASCII 編碼順序進行排序
+ key=str.lower
     + 將 list 內元素轉換為小寫，再依照 ASCII 編碼順序進行排序

### sort vs sorted 

```python
#小寫 > 大寫  > 數字
#字典序
#預設小到大
list1 = ['c',"B","a"]
print(list1)
list1.sort()
print(list1)

list2 = [2,6,1]
print(list2)
list3 = sorted(list2)
print(list2)
print(list3)
```

## 資料反向
+ 可利用 reverse( ) 函數進行反向排序動作。
+ 這函數沒有傳回值，list 執行後就會進行反向排序。

```python
list4 = [5,9,2,4,3]
list4.sort()
print(list4)
list4.reverse()#大到小
print(list4)

list1 = ['c',"B","a"]
list5 = sorted(list1,reverse=True)
print(list5)
```

```python
list3 = ["a","B","C","d"]
list4 = sorted(list3)
print(list4)
list4 = sorted(list3,key=str.upper)#內容轉成大小排序
print(list4)
```

## 自訂函數

## Function

+ 自訂函數的目的：
  + 減少相同的語法重複出現。
  + 避免重複出現的目的，是未來維護程式只要維護一個地方，不用改好幾個地方。改好幾個地方代表出錯的機率愈大。
+ 被呼叫的函數必須放在程式區塊的上方，代表已經準備好了，下方的程式區塊才可以呼叫使用。
+ 自訂函數可以放在其他檔案內，之後以模組引用 import 的方式使用：
  + 那自訂函數的內容就交給其他人負責，而你就只負責主程式的流程。
+ 在python函數是一種物件

```python
def func1():
    print("Func1")
func1()
```

```python
def func2(x):
    print(x ** 2)
for i in range(1,4): 
    func2(i)    
```

## function回傳

```python
def func3(v1,v2):
    v3 = v1 * v2
    return v3
ans = func3(2,5)
print(ans)
```

```python
def func4(v1,v2,v3):
    return v1+2,v2+3,v3+6
x,y,z = func4(2,3,5)
print(x,y,z)
```

```python
def func_sum(a,b):
    c = a + b
    return c
z = func_sum(10,15)
print(z)
```

### 查看回傳值數量與數值

```python
for i in func4(2,3,4):
    print(i)
rLen =  len(func4(2,3,4))    
print(rLen)
```

## 變數影響範圍
+ 函數外的變數：
  + 函數內可以顯示該變數內容
  + 不屬於函數的區域內都可以使用
+ 函數內的變數：
  + 只在函數內產生效果，不會影響函數外的變數
  + 若函數內沒有進行變數宣告而進行改變內容動作將會產生錯誤訊息!

```python
a = 5
def func_sum():
    a = 10
    print("函數內:",a)
func_sum()    
```

```python
a = 5
def func_sum():
    a = 10
    print("函數內:",a)
print("a:",a)
func_sum()  
print("a:",a)
```

```python
a = 5
def func_sum():
    print("函數內:",a)
func_sum()    
```

```python
a = 5
def func_sum():
    a = a + 1 #因為func_sum 沒有宣告a執行時出錯
    print("函數內:",a)
print("函數外1",a)
func_sum()
print("函數外2",a)
```

```python
a = 5
def func_sum():
    global a #宣告a是外部變數
    a = a + 1 #因為func_sum 沒有宣告a執行時出錯
    print("函數內:",a)
print("函數外1",a)
func_sum()
print("函數外2",a)
```
