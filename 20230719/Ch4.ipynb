## replace 可以將指定的字串進行更換：

str.replace(舊字串, 新字串)

replace 可以指定您要第幾個位置後不進行更換：

str.replace(舊字串, 新字串,幾次)

位置編號由 0 開始


```python
str1="This is Python,That is Java;This is SQLite,That is MySQL"
print("原本的字串:",str1)
str2 = str1.replace("is","-")
print("is 改為-的字串:",str2)
str3 = str1.replace("is","-",2)
print("is 改為-的字串:",str3)
```

```python
import re
str1="This is Python,That is Java;This is SQLite,That is MySQL"
p=r"\bis\b"
result = re.sub(p,"-",str1)
print(result)
```

## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。


```python
str1 = "this is Python Tutorial, there"
search1 = "Python"
print(str1.find(search1))
#刪除某個字
inx = str1.find(search1)
newStr = str1[:inx]+str1[inx+len(search1):]
print(newStr)
```

```python
str1 = "this is Python Tutorial,there"
s2 = "not"
inx = str1.find(s2)
print(inx)
s3 = "t"
inx = str1.find(s3)
print(inx)
inx = str1.find(s3,4)#由index:4開始找
print(inx)
inx = str1.find(s3,11,20)
print(inx)
```

```python
str1 = "this is Python Tutorial,there"
s3 = "t"
inx = 0
while True:
    inx = str1.find(s3,inx)
    if inx == -1:
        break
    print(inx)    
    inx += 1     
    
        
```
