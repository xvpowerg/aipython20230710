## 如何知道變數的資料型態？
可使用 type( ) 指令顯示。

變數內容變更後會儲存於不同的記憶體位址。那要如何查詢位址呢？

可使用 id( ) 指令顯示。


```python
var1 = 20
print(var1)
print(type(var1))
var2 = 123.45
print(type(var2))
var3 = True
print(type(var3))
var4 = 'string1'
print(type(var4))
```

```python
var5 = 10
print(id(var5))
var5 = 12.5
print(id(var5))
var5 = 10
print(id(var5))
```

```python
var6 = 1.57
print(type(var6))
print(var6.__class__)
```

```python
a,b,c = 10,3,1.56
print(a,b,c)
```
